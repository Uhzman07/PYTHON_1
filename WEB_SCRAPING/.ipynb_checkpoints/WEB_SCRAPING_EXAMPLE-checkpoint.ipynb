```python
import requests
import bs4
```

```python
# We can create a base url that will be changed
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
```

```python
res = requests.get(base_url.format(1))
```

```python
soup = bs4.BeautifulSoup(res.text, 'lxml')
```

```python
soup.select(".product_pod")
```

```python
len(soup.select(".product_pod"))
```

```python
products = soup.select(".product_pod")
```

```python
example = products[0]
```

```python
example.select(".star-rating.Three") # If there is a space inbetween the class name, we need to make use of a dot to replace it
```

```python
example.select(".star-rating.Two")
```

```python
example.select('a')[1]['title']
```

```python
# We can check if something is 2 stars (starting call in, example.select(rating))
# example.select('a')[1]['title'] to grab the book title
```

```python
two_star_titles = []

for n in range(1,51):

    scrape_url = base_url.format(n)

    res = requests.get(scrape_url)

    soup = bs4.BeautifulSoup(res.text, 'lxml')
    books = soup.select(".product_pod")

    for book in books:
        if len(book.select('.star-rating.Two')) != 0:
            book_title = book.select('a')[1]['title']
            two_star_titles.append(book_title)
```

```python
# Inspect the site and use Beautiful Soup to extract the top ten tags from the requests
# That is, the one shown on the to right from the home page(e.g. Love, Inspirational, Life, e.t.c)
```

```python
# we could have multiple pages
```

```python
url = 'http://quotes.toscrape.com/page/'
```

```python
authors = set()

for page in range(1,10):
    page_url = url + str(page)

    res = requests.get(page_url)

    soup = bs4.BeautifulSoup(res.text, "lxml")

    for name in soup.select(".author"):
        authors.add(name.text)
```

```python
page_still_valid = True
authors = set()
page = 1

while page_still_valid:
    page_url = url + str(page)

    res = requests.get(page_url)

    # We need to check if it is a valid page

    if "No quotes found!" in res.text:
        break

    soup = bs4.BeautifulSoup(res.text, "lxml")

    for name in soup.select(".author"):
        authors.add(name.text)

    page = page+1
```
